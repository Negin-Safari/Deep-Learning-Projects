# **McCulloch_Pitts Neural Network**

```python
import numpy as np
```

## **McCulloch_Pitts_Neuron**

```python
class McCulloch_Pitts_Neuron():

    def __init__ (self, weights, threshold):
        self.weights = weights
        self.threshold = threshold

    def model (self, x):
        if self.weights @ x >= self.threshold:
            return 1
        else:
            return 0
```

## **1-2) 1 Layer NN**

```python
def seven_seg_id_1L ():
    
    neur1 = McCulloch_Pitts_Neuron ([1, -1, 1, 1, 1, 1, 1], 6)
    neur2 = McCulloch_Pitts_Neuron ([1, 1, 1, -1, -1, -1, -1], 3)
    neur3 = McCulloch_Pitts_Neuron ([1, 1, 1, 1, 1, 1, 1], 7)
    neur4 = McCulloch_Pitts_Neuron ([1, 1, -1, 1, 1, 1, 1], 6)

    o1 = neur1.model(np.array([1, 0, 1, 1, 1, 1, 1]))
    o2 = neur2.model(np.array([1, 0, 1, 1, 1, 1, 1]))
    o3 = neur3.model(np.array([1, 0, 1, 1, 1, 1, 1]))
    o4 = neur4.model(np.array([1, 0, 1, 1, 1, 1, 1]))
    
    print ("test1 : 1011111")
    print ("o1 o2 o3 o4")
    print (o1, o2, o3, o4)

    o1 = neur1.model(np.array([1, 1, 1, 0, 0, 0, 0]))
    o2 = neur2.model(np.array([1, 1, 1, 0, 0, 0, 0]))
    o3 = neur3.model(np.array([1, 1, 1, 0, 0, 0, 0]))
    o4 = neur4.model(np.array([1, 1, 1, 0, 0, 0, 0]))
    
    print ("test2 : 1110000")
    print ("o1 o2 o3 o4")
    print (o1, o2, o3, o4)

    o1 = neur1.model(np.array([1, 1, 1, 1, 1, 1, 1]))
    o2 = neur2.model(np.array([1, 1, 1, 1, 1, 1, 1]))
    o3 = neur3.model(np.array([1, 1, 1, 1, 1, 1, 1]))
    o4 = neur4.model(np.array([1, 1, 1, 1, 1, 1, 1]))
    
    print ("test3 : 1111111")
    print ("o1 o2 o3 o4")
    print (o1, o2, o3, o4)

    o1 = neur1.model(np.array([1, 1, 0, 1, 1, 1, 1]))
    o2 = neur2.model(np.array([1, 1, 0, 1, 1, 1, 1]))
    o3 = neur3.model(np.array([1, 1, 0, 1, 1, 1, 1]))
    o4 = neur4.model(np.array([1, 1, 0, 1, 1, 1, 1]))
    
    print ("test4 : 1101111")
    print ("o1 o2 o3 o4")
    print (o1, o2, o3, o4)
```

```python
seven_seg_id_1L()
```
